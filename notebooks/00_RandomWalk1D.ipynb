# Random Walk 1D
This notebook simulates a random walk in one dimension.

The user can define the number of steps, and the cumulative positions are displayed.

```python
# Python code for 1-D random walk (interactive)
import numpy as np
import matplotlib.pyplot as plt

# Ask the user for the number of steps
try:
    n = int(input("Enter the number of steps for the random walk: "))
    if n <= 0:
        raise ValueError
except ValueError:
    print("Number invalid, n=1000 will be used by default.")
    n = 1000

# Define initial position
start = 0
positions = np.zeros(n)
positions[0] = start

# Generate random steps [-1, 0, 1]
values = np.random.choice([-1, 0, 1], size=n)

# Calculated accumulated positions
for i in range(1, n):
    positions[i] = positions[i-1] + values[i]

# Graphic the 1D random walk
plt.figure(figsize=(10,5))
plt.plot(positions, color="blue", linewidth=2, alpha=0.7)
plt.title(f"Random Walk 1D ({n} steps)")
plt.xlabel("Number of steps")
plt.ylabel("Positions")
plt.grid(alpha=0.3)
plt.show()
```

## Conclusions

- The walker's final position depends entirely on random steps.
- The walk may remain close to the initial position or deviate significantly, reflecting the stochastic nature of the walk.
- This simple model is the basis for understanding more complex processes such as Brownian motion or financial models.

```python
print(f"Final position: {positions[-1]}")
print(f"Maximum position reached: {positions.max()}")
print(f"Minimum position reached: {positions.min()}")
print(f"Standard deviation of positions: {positions.std():.2f}")

plt.hist(positions, bins=20, color="skyblue", edgecolor="black")
plt.title("Distribution of positions during the walk")
plt.xlabel("Position")
plt.ylabel("Frequency")
plt.show()
```
