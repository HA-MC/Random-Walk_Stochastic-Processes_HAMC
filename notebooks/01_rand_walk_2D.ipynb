# Random Walk 2D

## Simulation Paramaters

- `n`: Number of steps in the random walk
- `x`, `y`: Arrays that store the coordinates on the plane
- `val`: Random variable to choose the direction

## 2D Random Step Generation

Each step is chosen uniformly from four possible directions:
1. Right (x + 1)
2. Left (x - 1)
3. Up (y + 1)
4. Down (y - 1)

## Random walk visualization

- The trajectory `(x, y)` is graphed on a Cartesian plane.
- A copy PDF is saved for documentation/reporting.

## Observations

- The trajectory shows random dispersion typical of a stochastic process
- The greater the number of steps, the further the walks moves from the origin.
- This model serves as the basis for more complex processes such as Brownian motion.

```python
import numpy
import pylab
import random

# defining the number of steps
n = 10000

#creating two array for containing x and y coordinate
#of size equals to the number of size and filled up with 0's
x = numpy.zeros(n)
y = numpy.zeros(n)

# filling the coordinates with random variables
for i in range(1, n):
    val = random.randint(1, 4)
    if val == 1:
        x[i] = x[i - 1] + 1
        y[i] = y[i - 1]
    elif val == 2:
        x[i] = x[i - 1] - 1
        y[i] = y[i - 1]
    elif val == 3:
        x[i] = x[i - 1]
        y[i] = y[i - 1] + 1
    else:
        x[i] = x[i - 1]
        y[i] = y[i - 1] - 1


# plotting stuff:
pylab.title("Random Walk 2D grid ($n = " + str(n) + "$ steps)")
pylab.plot(x, y)
pylab.savefig("plots/01_rand_walk_grid_2D.png",bbox_inches="tight",dpi=600)
pylab.show()
```

## Conclusions of the 2D Random Walk

- The two dimensional random walk exhibits inherently stochastic behavior, characterized by trajectories without a deterministic pattern.
- The particle's movement exhibits a discrete diffusion phenomenon, where the distance to the initial point increases as the number of steps increases.
- The simulation qualitatively reproduces features of Brownnian motion, constituing a simplified model of diffusion processes in physical systems.
- The distance behavior, the code does not explicitly calculate it, but the average displacement of the particle from the start point is zero, since movements are equally probably in all directions. However the mean squared displacement (MSD) from the origin increases linearly zith the number of steps (r^2=cN), where N is the number of the steps and C is a constant. This is a key charasteristic of diffusive processes and demostrates that, even though the average displacement is zero, the particle moves away from the origin over time.
- Origin Recurrence: In a 2D random walk, there is a high probability that the particle will return to the origin at same point. Unlike a 3D random walk, where the probability of returning to the starting point is less than 1, the 2D random walk is classified as a recurrent process, meaning that the particle will return to the origin with a probability of 1 (almost certain).
-Position Distribution: The distributions of steps of walker's positions (in both X and Y coordinates) tends to follow a normal (Gaussian) distribution. This is due to the central Limit Theorem, which states that the sum of a large number of independent random variables (in this case, steps) approximates a normal distribution. This implies that the particle is more likely to be found near the origin at very distant points, and the positions qre dispersed in a circular pattern around (0,0);
