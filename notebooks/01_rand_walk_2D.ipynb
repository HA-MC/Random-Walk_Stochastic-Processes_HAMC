# Random Walk 2D

## Simulation Paramaters

- `n`: Number of steps in the random walk
- `x`, `y`: Arrays that store the coordinates on the plane
- `val`: Random variable to choose the direction

## 2D Random Step Generation

Each step is chosen uniformly from four possible directions:
1. Right (x + 1)
2. Left (x - 1)
3. Up (y + 1)
4. Down (y - 1)

## Random walk visualization

- The trajectory `(x, y)` is graphed on a Cartesian plane.
- A copy PDF is saved for documentation/reporting.

## Observations

- The trajectory shows random dispersion typical of a stochastic process
- The greater the number of steps, the further the walks moves from the origin.
- This model serves as the basis for more complex processes such as Brownian motion.

```python
import numpy
import pylab
import random

# defining the number of steps
n = 10000

#creating two array for containing x and y coordinate
#of size equals to the number of size and filled up with 0's
x = numpy.zeros(n)
y = numpy.zeros(n)

# filling the coordinates with random variables
for i in range(1, n):
    val = random.randint(1, 4)
    if val == 1:
        x[i] = x[i - 1] + 1
        y[i] = y[i - 1]
    elif val == 2:
        x[i] = x[i - 1] - 1
        y[i] = y[i - 1]
    elif val == 3:
        x[i] = x[i - 1]
        y[i] = y[i - 1] + 1
    else:
        x[i] = x[i - 1]
        y[i] = y[i - 1] - 1


# plotting stuff:
pylab.title("Random Walk 2D grid ($n = " + str(n) + "$ steps)")
pylab.plot(x, y)
pylab.savefig("plots/01_rand_walk_grid_2D.png",bbox_inches="tight",dpi=600)
pylab.show()
```
